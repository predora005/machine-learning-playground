```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from solver import KMeans
```

```python
np.random.seed(0)
n = 100
clusters = 3

points1 = np.random.randn(n, 2)
points2 = np.random.randn(n, 2) + np.array([3, 0])
points3 = np.random.randn(n, 2) + np.array([3, 3])

points = np.r_[points1, points2, points3]
np.random.shuffle(points)

model = KMeans(clusters)
model.fit(points)
```

```python
markers = ["+", "*", "o", "^"]
colors = ["r", "g", "b", "m"]
for i in range(clusters):
    p = points[model.labels_ == i, :]
    plt.scatter(p[:, 0], p[:, 1], color=colors[i], marker=markers[i])

plt.show()
```
