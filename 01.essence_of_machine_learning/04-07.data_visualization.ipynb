# データの可視化

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 折れ線グラフ

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 7, 4, 8])
plt.plot(x, y, color="r")
plt.show()
```

## 散布図

```python
plt.scatter(x, y, color="r")
plt.show()
```

## 曲線のグラフ

```python
x = np.linspace(-5, 5, 50)
y = x**2
plt.plot(x, y, color="r")
plt.show()
```

## 複数の線を表示する

```python
y2 = (x - 2)**2
plt.plot(x, y, color="r")
plt.plot(x, y2, color="k", linestyle="--")
plt.show()
```

## ヒストグラム

```python
# サイコロを10回振った時の合計
np.random.seed(0)
l = []
for _ in range(1000):
    l.append(np.random.randint(1, 6+1, size=10).sum())
    
plt.hist(l, 20, color="gray")#20はビンの数
plt.show()
```

## 複数のグラフを並べて表示する

```python
x = np.linspace(-5, 5, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)

fig, axes = plt.subplots(2, 1)
axes[0].set_ylim([-1.5, 1.5])
axes[1].set_ylim([-1.5, 1.5])
axes[0].plot(x, sin_x, color="r")
axes[1].plot(x, cos_x, color="k")

plt.show()
```

## 等高線の描画

```python
def f(x, y):
    return x**2 + y**2 / 4

x = np.linspace(-10, 10, 101)
y = np.linspace(-10, 10, 101)
xmesh, ymesh = np.meshgrid(x, y)
z = f(xmesh.ravel(), ymesh.ravel()).reshape(xmesh.shape)

#print(xmesh, ymesh)
#print(xmesh.shape)

fig, axes = plt.subplots()
axes.contour(x, y, z, colors="k", levels=[1, 2, 3, 4, 5, 8, 16, 32, 64])
#axes.set_xlim([-10, 10])
#axes.set_ylim([-10, 10])
plt.show()
```

```python
colors = ["0.1", "0.3", "0.5", "0.7"]
levels = [1, 2, 4, 8, 16, 32, 64]
plt.contourf(x, y, z, colors=colors, levels=levels)
plt.show()
```
